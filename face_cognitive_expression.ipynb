```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
import sys
import os
from random import shuffle


# path for images
'''
Having 6 different emotions to identify
Configuring the folders for each emotion
Configuring the labels for each emotion
Training data to put in 6 subfolders, there folder name correponds to label

'''
TRAINING_FOLDER = '/content/CK+48/train/'
TESTING_FOLDER = '/content/CK+48/test/'
HAPPY_FOLDER = 'happy'
SAD_FOLDER = 'sad'
#FEAR_FOLDER = 'fear'
#ANGRY_FOLDER = 'angry'
#SURPRISE_FOLDER = 'surprise'
NEUTRAL_FOLDER = 'neutral'

#labels for images
label_Happy = 'Happy'
label_Sad = 'Sad'
#label_Fear = 'Fear'
#label_Angry = 'Angry'
#label_Surprise = 'Surprise'
label_Neutral = 'Neutral'

# function to process image
# given a image name in same folder as in this program, it outputs processed image
'''
function to process one image
takes full path of image as argument
image name should be ending with jpg/png/jpeg
creates a numpy array of image after reading and converting to GRAYSCALE and resize
'''

row = 90
cox = 90

def process(imageName):
    output = None
    #check the format of image
    if imageName.endswith('jpg') or imageName.endswith('png') or imageName.endswith('jpeg'):
        # read the image in grayscale
        grayscaleImage = cv2.imread(imageName,cv2.IMREAD_GRAYSCALE) 

        grayscaleImage = grayscaleImage/255.0

        # resize the image 28 * 28
        output = cv2.resize(grayscaleImage, (row,cox))
        

    #return the result
    return output

# contains training dataset images and labels (global)
training_dataset = []
testing_dataset = []
testing_label = []
training_label = []

# create training dataset with label
'''
function to create training dataset and associate it will corresponding label
iterate over each folder of 6 emotions
process each image using above function
and finally adds the processed image to training dataset
'''
def create_training_dataset():

    # happy emotions
    for imgName in os.listdir(TRAINING_FOLDER + HAPPY_FOLDER):
        if not imgName.startswith('.'):
            # call image processing method to get appropriate image
            path = TRAINING_FOLDER + HAPPY_FOLDER + "/" + imgName
            print("Processing.... ", path)
            #we are calling this function def process(imageName):
            #from here to process the image to convert it to the desire format (grayscale)
            img = process(path)
            # add to training dataset with appropriate label
            training_dataset.append(np.array(img))
            training_label.append(np.array(label_Happy))

    # sad emotions
    for imgName in os.listdir(TRAINING_FOLDER + SAD_FOLDER):
        if not imgName.startswith('.'):
            # call image processing method to get appropriate image
            path = TRAINING_FOLDER + SAD_FOLDER + "/" + imgName
            print("Processing.... ", path)
            img = process(path)
            # add to training dataset with appropriate label
            training_dataset.append(np.array(img))
            training_label.append(np.array(label_Sad))
        
        # training_dataset.append([np.array(img),np.array(label_Sad)])
    '''
    # fear emotions
    for imgName in os.listdir(TRAINING_FOLDER + FEAR_FOLDER):
        if not imgName.startswith('.'):
        # call image processing method to get appropriate image
            path = TRAINING_FOLDER + FEAR_FOLDER + "/" + imgName
            print("Processing.... ", path)
            img = process(path)
            # add to training dataset with appropriate label
            training_dataset.append(np.array(img))
            training_label.append(np.array(label_Fear))
        # training_dataset.append([np.array(img),np.array(label_Fear)])
    
    # surpise emotions
    for imgName in os.listdir(TRAINING_FOLDER + SURPRISE_FOLDER):
        if not imgName.startswith('.'):
            # call image processing method to get appropriate image
            path = TRAINING_FOLDER + SURPRISE_FOLDER + "/" + imgName
            print("Processing.... ", path)
            img = process(path)
            # add to training dataset with appropriate label
            training_dataset.append(np.array(img))
            training_label.append(np.array(label_Surprise))
        # training_dataset.append([np.array(img),np.array(label_Surprise)])

    # angry emotions
    for imgName in os.listdir(TRAINING_FOLDER + ANGRY_FOLDER):
        if not imgName.startswith('.'):
            # call image processing method to get appropriate image
            path = TRAINING_FOLDER + ANGRY_FOLDER + "/" + imgName
            print("Processing.... ", path)
            img = process(path)
            # add to training dataset with appropriate label
            training_dataset.append(np.array(img))
            training_label.append(np.array(label_Angry))
        # training_dataset.append([np.array(img),np.array(label_Angry)])
    ''' 
    # neutral emotions
    for imgName in os.listdir(TRAINING_FOLDER + NEUTRAL_FOLDER):
        if not imgName.startswith('.'):
        # call image processing method to get appropriate image
            path = TRAINING_FOLDER + NEUTRAL_FOLDER + "/" + imgName
            print("Processing.... ", path)
            img = process(path)
            # add to training dataset with appropriate label
            training_dataset.append(np.array(img))
            training_label.append(np.array(label_Neutral))
        # training_dataset.append([np.array(img),np.array(label_Neutral)])

    return training_dataset, training_label
# process testing dataset 
'''
function to process the test dataset
iterate over images in test dataset
process each image using process(image)
add it to test dataset
'''


def process_test_dataset():

    # happy emotions
    for imgName in os.listdir(TESTING_FOLDER  + HAPPY_FOLDER):
        if not imgName.startswith('.'):
            # call image processing method to get appropriate image
            path = TESTING_FOLDER  + HAPPY_FOLDER + "/" + imgName
            print("Processing.... ", path)
            #we are calling this function def process(imageName):
            #from here to process the image to convert it to the desire format (grayscale)
            img = process(path)
            # add to training dataset with appropriate label
            testing_dataset.append(np.array(img))
            testing_label.append(np.array(label_Happy))
    
    # sad emotions
    for imgName in os.listdir(TESTING_FOLDER  + SAD_FOLDER):
        if not imgName.startswith('.'):
            # call image processing method to get appropriate image
            path = TESTING_FOLDER  + SAD_FOLDER + "/" + imgName
            print("Processing.... ", path)
            img = process(path)
            # add to training dataset with appropriate label
            testing_dataset.append(np.array(img))
            testing_label.append(np.array(label_Sad))
        
        # training_dataset.append([np.array(img),np.array(label_Sad)])
    '''
    # fear emotions
    for imgName in os.listdir(TESTING_FOLDER  + FEAR_FOLDER):
        if not imgName.startswith('.'):
            # call image processing method to get appropriate image
            path = TESTING_FOLDER  + FEAR_FOLDER + "/" + imgName
            print("Processing.... ", path)
            img = process(path)
            # add to training dataset with appropriate label
            testing_dataset.append(np.array(img))
            testing_label.append(np.array(label_Fear))
        # training_dataset.append([np.array(img),np.array(label_Fear)])

    # surpise emotions
    for imgName in os.listdir(TESTING_FOLDER  + SURPRISE_FOLDER):
        if not imgName.startswith('.'):
            # call image processing method to get appropriate image
            path = TESTING_FOLDER  + SURPRISE_FOLDER + "/" + imgName
            print("Processing.... ", path)
            img = process(path)
            # add to training dataset with appropriate label
            testing_dataset.append(np.array(img))
            testing_label.append(np.array(label_Surprise))
        # training_dataset.append([np.array(img),np.array(label_Surprise)])

    # angry emotions
    for imgName in os.listdir(TESTING_FOLDER  + ANGRY_FOLDER):
        if not imgName.startswith('.'):
            # call image processing method to get appropriate image
            path = TESTING_FOLDER  + ANGRY_FOLDER + "/" + imgName
            print("Processing.... ", path)
            img = process(path)
            # add to training dataset with appropriate label
            testing_dataset.append(np.array(img))
            testing_label.append(np.array(label_Angry))
        # training_dataset.append([np.array(img),np.array(label_Angry)])
    '''    
    # neutral emotions
    for imgName in os.listdir(TESTING_FOLDER  + NEUTRAL_FOLDER):
        if not imgName.startswith('.'):
            # call image processing method to get appropriate image
            path = TESTING_FOLDER  + NEUTRAL_FOLDER + "/" + imgName
            print("Processing.... ", path)
            img = process(path)
            # add to training dataset with appropriate label
            testing_dataset.append(np.array(img))
            testing_label.append(np.array(label_Neutral))
        # training_dataset.append([np.array(img),np.array(label_Neutral)])

    return testing_dataset, testing_label


##################
train_dateset, train_label = create_training_dataset()
test_dataset, test_label = process_test_dataset()


```

```python
######################
#Splitting the feature and label for train dataset
empty_arr=[i for i,x in enumerate(train_dateset) if x ==[]]
feature_train = [i for i in train_dateset if i != []]
feature_train = np.array(feature_train).reshape(-1, row, cox, 1)
feature_train =np.where(np.isnan(feature_train), 0, feature_train)
#feature_train = feature_train.fillna(0)

labels_train = [i for i in train_label]
labels_train = np.array(labels_train).astype(str)


#Splitting the feature and label for test dataset
# feature_test = [i for i in test_dataset]
feature_test = [i for i in test_dataset if i != []]
feature_test = np.array(feature_test).reshape(-1, row, cox, 1)
feature_test =np.where(np.isnan(feature_test), 0, feature_test)

labels_test = [i for i in test_label]



from sklearn import preprocessing
number = preprocessing.LabelEncoder() 

labels_train= number.fit_transform(labels_train)
labels_test= number.fit_transform(labels_test)


######################
import keras
from keras.models import Sequential 
from keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
from keras.layers import Dropout, Input, BatchNormalization
from sklearn.metrics import confusion_matrix, accuracy_score 
from keras.losses import categorical_crossentropy
from keras.optimizers import Adadelta 
from keras import regularizers

y_train = keras.utils.to_categorical(labels_train)
y_test = keras.utils.to_categorical(labels_test)

filters = 10
n_classes = y_train.shape[1]
def larger_model(): 

	# create model
    model = Sequential()
    model.add(Conv2D(filters, (3, 3), input_shape=(row, cox, 1), activation='relu'))
    model.add(BatchNormalization())
    model.add(Flatten())
    model.add(Dense(100, activation='relu'))
    model.add(Dense(n_classes, activation='softmax'))
# Compile model
    model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
    return model
# build the model
model = larger_model()
# Fit the model
model.fit(feature_train, y_train, validation_split=0.2, shuffle=True, epochs=10, batch_size=70)
# Final evaluation of the model
scores = model.evaluate(feature_test, y_test, verbose=0)
print("Large CNN Error: %.2f%%" % (100-scores[1]*100))
 
```

```python
from IPython.display import display, Javascript
from google.colab.output import eval_js
from base64 import b64decode

def take_photo(filename='photo.jpg', quality=0.8):
  js = Javascript('''
    async function takePhoto(quality) {
      const div = document.createElement('div');
      const capture = document.createElement('button');
      capture.textContent = 'Capture';
      div.appendChild(capture);

      const video = document.createElement('video');
      video.style.display = 'block';
      const stream = await navigator.mediaDevices.getUserMedia({video: true});

      document.body.appendChild(div);
      div.appendChild(video);
      video.srcObject = stream;
      await video.play();

      // Resize the output to fit the video element.
      google.colab.output.setIframeHeight(document.documentElement.scrollHeight, true);

      // Wait for Capture to be clicked.
      await new Promise((resolve) => capture.onclick = resolve);

      const canvas = document.createElement('canvas');
      canvas.width = video.videoWidth;
      canvas.height = video.videoHeight;
      canvas.getContext('2d').drawImage(video, 0, 0);
      stream.getVideoTracks()[0].stop();
      div.remove();
      return canvas.toDataURL('image/jpeg', quality);
    }
    ''')
  display(js)
  data = eval_js('takePhoto({})'.format(quality))
  binary = b64decode(data.split(',')[1])
  with open(filename, 'wb') as f:
    f.write(binary)
  return filename
```
